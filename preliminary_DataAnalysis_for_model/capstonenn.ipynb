```python
training_df = music_df.drop(columns=["artists", "track_name"])
```

```python
training_df.head()
```

```python
training_df.head()
```

```python
neigh = NN(n_neighbors=2, radius=0.4)
```

```python
training_df = training_df.drop(columns="track_id")
```

```python
training_df
```

```python
le = LabelEncoder()
le.fit(training_df["track_genre"])
training_df["track_genre"] = le.transform(training_df["track_genre"])
```

```python
norm = Normalizer()
#norm.fit(training_df["duration_ms"].values.reshape(-1,1))
#training_df["duration_ms"] = norm.transform(training_df["duration_ms"].values.reshape(-1,1))
training_df["duration_ms"] = preprocessing.StandardScaler().fit_transform(training_df["duration_ms"].values.reshape(-1,1))
```

```python
#norm = Normalizer()
#norm.fit(training_df["popularity"].values.reshape(-1,1))
training_df["popularity"] = preprocessing.StandardScaler().fit_transform(training_df["popularity"].values.reshape(-1,1))
```

```python
#norm = Normalizer()
#norm.fit(training_df["tempo"].values.reshape(-1,1))
training_df["tempo"] = preprocessing.StandardScaler().fit_transform(training_df["tempo"].values.reshape(-1,1))
```

```python
training_df["loudness"] = preprocessing.StandardScaler().fit_transform(training_df["loudness"].values.reshape(-1,1))
```

```python
training_df["acousticness"] = preprocessing.StandardScaler().fit_transform(training_df["acousticness"].values.reshape(-1,1))
training_df["instrumentalness"] = preprocessing.StandardScaler().fit_transform(training_df["instrumentalness"].values.reshape(-1,1))
training_df["time_signature"] = preprocessing.StandardScaler().fit_transform(training_df["time_signature"].values.reshape(-1,1))
training_df["mode"] = preprocessing.StandardScaler().fit_transform(training_df["mode"].values.reshape(-1,1))
training_df["key"] = preprocessing.StandardScaler().fit_transform(training_df["key"].values.reshape(-1,1))
training_df = training_df.drop(columns=["explicit"])
```
